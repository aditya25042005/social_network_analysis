```python

 from bs4 import BeautifulSoup
```

```python
HTMLFileToBeOpened = open("post_comments_1.html", "r")
contents = HTMLFileToBeOpened.read()
```

```python
beautifulSoupText = BeautifulSoup(contents, 'lxml')
beautifulSoupText
```

```python
import pandas as pd
```

```python
cells = beautifulSoupText.select("td._2pin._a6_q >div >div")
for cell in cells:
  print(cell)

```

```python
cells = beautifulSoupText.select("td._2pin._a6_q")
account_names = []
for cell in cells:
    # 2. Find the label text (like "Media Owner" or "Comment")
    #    that is DIRECTLY inside the <td>.
    cell_label = cell.find(text=True, recursive=False)

    if not cell_label:
        continue # Skip this cell if it has no label

    # ---
    # Solution A: If the label is "Media Owner"
    # ---
    if "Media Owner" in cell_label:
        div = cell.select_one("div > div")
        if div:
            if div.text.strip() =='adityakarn0001':
                continue
            account_names.append(div.text.strip())

    # ---
    # Solution B: If the label is "Comment"
    # ---
    elif "Comment" in cell_label:
        # Get the comment's text
        div = cell.select_one("div > div")
        if div:
            comment_text = div.text.strip()

            # Check if the comment starts with @
            if comment_text.startswith("@"):
                # Split the text at the first space
                parts = comment_text.split(" ", 1)

                # The first part is the name, so remove the "@"
                name = parts[0][1:]
                if name=='adityakarn0001':
                  continue

                account_names.append(name)

# This list will have ALL account names, from both
# "Media Owner" rows and "@" mentions.
print(account_names)
#list of people invloved in comment
```

```python
account_count={}

for name in account_names:
 account_count[name] = account_count.get(name, 0) + 1
print(account_count)
```

```python
from google.colab import files
df = pd.read_excel('followers.xlsx')
print(df.head())
df['comment'] = df['followers'].map(account_count)
df['comment'] = df['comment'].fillna(0).astype(int)
print(df.head())
df.to_excel("followers_comment.xlsx", index=False)
print("File saved successfully.")
files.download("followers_comment.xlsx")

```
