```python
 from bs4 import BeautifulSoup
```

```python
HTMLFileToBeOpened = open("followers_1.html", "r")
contents = HTMLFileToBeOpened.read()


```

```python
beautifulSoupText = BeautifulSoup(contents, 'lxml')
beautifulSoupText
```

```python
import pandas as pd
```

```python
links=beautifulSoupText.find_all('a')
followers=[]
for link in links:
   followers.append(link.text)
df = pd.DataFrame()
df['followers']=followers
df.to_excel('followers.xlsx', index = False)



```
